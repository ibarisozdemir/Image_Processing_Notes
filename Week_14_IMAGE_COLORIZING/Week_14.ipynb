```python
import numpy as np
import cv2
import os
from pathlib import Path
import torch
from deoldify import device
from deoldify.device_id import DeviceId
from deoldify.visualize import *
import warnings
warnings.filterwarnings("ignore")

```

```python
def colorize_image(input_path,output_path,render_factor=35):
    try:

        input_path=str(Path(input_path).absolute())
        output_path=str(Path(output_path).absolute())
        
        device.set(device=DeviceId.GPU0 if torch.cuda.is_available() else DeviceId.CPU)

        print("Model Yükleniyor...")

        colorizer=get_image_colorizer(artistic=True)
        
        print ("Reklendirme İşlemi Başlatılıyor ....")

        colorizer.plot_transformed_image(path=input_path,render_factor=35,compare=True)
        
        return True
    except Exception as e: 
        print(f"Hata oluştu.Hata = {e}")
        return False


my_input_img="input_path.jpg"
my_output_img="my_colorized_img.png"

colorize_image(my_input_img,my_output_img,render_factor=40)
    
```
